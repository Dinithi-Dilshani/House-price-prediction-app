```python
data = data.drop(['id', 'Date'], axis=1)
```

Handle outliers

```python
Q1 = data['Price'].quantile(0.25)
Q3 = data['Price'].quantile(0.75)
IQR = Q3 - Q1

data = data[(data['Price'] >= Q1 - 1.5*IQR) & (data['Price'] <= Q3 + 1.5*IQR)]
```

```python
sns.boxenplot(data['Price'])
```

```python
data = data.drop_duplicates()
```

Define Features (X) and Target (y)

```python
x = data.drop('Price', axis=1)
y = data['Price']
```

Scale Numerical Features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
```

Train–Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
import numpy as np

from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
```

Define a function to evaluate models

```python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    return rmse, r2
```

```python
results = {}
```

LINEAR REGRESSION

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
rmse, r2 = evaluate_model(lr, x_test, y_test)
results["Linear Regression"] = {"model": lr, "rmse": rmse, "r2": r2}
```

RANDOM FOREST REGRESSOR

```python
rf_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
}

rf_grid = GridSearchCV(
    RandomForestRegressor(random_state=42),
    rf_params,
    cv=3,
    scoring='r2',
    n_jobs=-1
)

rf_grid.fit(x_train, y_train)
best_rf = rf_grid.best_estimator_
rmse, r2 = evaluate_model(best_rf, x_test, y_test)
results["Random Forest (Tuned)"] = {"model": best_rf, "rmse": rmse, "r2": r2}
```

GRADIENT BOOSTING REGRESSOR

```python
gb_params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 4]
}

gb_grid = GridSearchCV(
    GradientBoostingRegressor(random_state=42),
    gb_params,
    cv=3,
    scoring='r2',
    n_jobs=-1
)

gb_grid.fit(x_train, y_train)
best_gb = gb_grid.best_estimator_
rmse, r2 = evaluate_model(best_gb, x_test, y_test)
results["Gradient Boosting (Tuned)"] = {"model": best_gb, "rmse": rmse, "r2": r2}
```

SUPPORT VECTOR REGRESSOR

```python
svr_params = {
    'C': [1, 10, 50],
    'epsilon': [0.1, 0.2],
    'kernel': ['rbf']
}

svr_grid = GridSearchCV(
    SVR(),
    svr_params,
    cv=3,
    scoring='r2',
    n_jobs=-1
)

svr_grid.fit(x_train, y_train)
best_svr = svr_grid.best_estimator_
rmse, r2 = evaluate_model(best_svr, x_test, y_test)
results["SVR (Tuned)"] = {"model": best_svr, "rmse": rmse, "r2": r2}
```

Display Result

```python
print("\nMODEL PERFORMANCE COMPARISON:")
print("--------------------------------------")

for model_name, info in results.items():
    print(f"{model_name} → RMSE: {info['rmse']:.2f}, R2 Score: {info['r2']:.4f}")

best_model_name = max(results, key=lambda m: results[m]["r2"])
best_model = results[best_model_name]["model"]

print("\n\nBEST MODEL IS:", best_model_name)
```

Model Comparison Bar Charts

```python
model_names = list(results.keys())
rmse_values = [results[m]['rmse'] for m in model_names]
r2_values = [results[m]['r2'] for m in model_names]


# -----------------------------
# Plot RMSE Comparison
# -----------------------------
plt.figure(figsize=(10, 5))
plt.bar(model_names, rmse_values)
plt.title("Model Comparison - RMSE")
plt.ylabel("RMSE")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -----------------------------
# Plot R2 Score Comparison
# -----------------------------
plt.figure(figsize=(10, 5))
plt.bar(model_names, r2_values)
plt.title("Model Comparison - R2 Score")
plt.ylabel("R2 Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
